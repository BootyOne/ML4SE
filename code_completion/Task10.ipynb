1. [Copilot](https://github.com/features/copilot)

Что такое Copilot?
Генеральный директор GitHub Нэт Фриндман в своём блоге говорит о разработке просто:

«GitHub Copilot — ваш искусственный напарник-программист»

Самая интересная способность “Copilot” — превращать обычные человеческие слова и предложения в программный код. Расширение работает на базе нейросети Codex компании OpenAI, которая способна переводить английский язык в язык программирования. По крайней мере, так кажется пользователю.

В последнем исследовании разработчики сервиса обнаружили, что пользователи приняли в среднем 26% всех подсказок, показанных GitHub Copilot. Также было подсчитано, что GitHub Copilot создает в среднем более 27% файлов кода разработчиков, а на некоторых языках, таких как Python, этот показатель достигает 40%. Однако GitHub Copilot не пишет идеальный код. Он предназначен для создания наилучшего возможного кода с учетом контекста, к которому у него есть доступ, но он не проверяет предлагаемый код, поэтому программа не всегда может работать без дополнительного вмешательства человека или вовсе иметь смысл… Для предлагаемого кода некоторые языки, такие как Python, JavaScript, TypeScript и Go, могут работать лучше, чем другие языки программирования, ведь база для нейросети формируется из анализа разработок других людей, а другие люди пишут на том, что имеет спрос и популярность. Как и любой другой код, варианты, предложенные GitHub Copilot, должны быть тщательно протестированы и проверены. Как разработчик, вы всегда несете ответственность.

3. [Captain Stack](https://github.com/hieunc229/copilot-clone)

Captain Stack - продукт для предложение по коду для VSCode

Этот продукт в чем-то похож на предложение кода Github Copilot. Но вместо использования искусственного интеллекта он отправляет ваш поисковый запрос в Google, затем извлекает ответы StackOverflow и Github Gist и автоматически заполняет их для вас.

Чтобы запустить встроенное завершение, вам нужно будет ввести:

//find {ваше ключевое слово}. (начинайте с //find, заканчивайте точкой .)

Например:

//find binary search in javascript.

IntelliCode - это экспериментальный помощник, обученный на подмножестве проектов GitHub, продукта Microsoft, доступного только для Visual Studio. Одной из наиболее привлекательных особенностей IntelliCode является командное завершение. Командное завершение может оказаться полезным для организаций с архитектурой на базе Microsoft и разработчиков, привыкших к Visual Studio в эпоху удаленной работы.

IntelliCode помещает наиболее полезные методы в начало списка завершения, тем самым экономя ваше время. Возможности IntelliCode не ограничиваются завершением операторов. Справка по сигнатурам также дает рекомендации по перегрузке метода, наиболее подходящей в данной ситуации.

Завершения команд

IntelliCode может предоставлять рекомендации по коду и отправлять их участникам вашей команды. В этой предварительной версии функции вы можете построить командную модель, чтобы давать рекомендации по коду, которого нет в открытом доступе, например методов из ваших собственных служебных классов или вызовов библиотек для конкретных доменов. Интегрируйте собственные задачи сборки в конвейер, чтобы завершение кода для команды выполнялось в соответствии с последними изменениями в репозитории.

Дополнение аргументов

Помимо завершения сигнатур операторов, IntelliCode также рекомендует аргументы, помогая быстро выбрать подходящий аргумент.

IntelliCode устанавливается по умолчанию с любой рабочей нагрузкой, которая поддерживает C#, C++, TypeScript/JavaScript или XAML в Visual Studio 2022.

5. [Code Whisperer](https://aws.amazon.com/ru/codewhisperer/) от Amazon

Поскольку Microsoft соблазняла разработчиков через GitHub и Copilot, было ясно, что в дело вступят другие технологические гиганты. В июне 2022 года Amazon запустила собственный инструмент для парного программирования ИИ под названием CodeWhisperer. По словам Amazon, CodeWhisperer был обучен «на миллиардах строк кода, взятых из репозиториев с открытым исходным кодом, внутренних репозиториев Amazon, документации API и форумов».


На момент написания CodeWhisperer доступен в предварительной версии только после регистрации и может быть установлен в любой IDE, которая поддерживает инструменты AWS IDE.

Поддерживаемые языки: C#, Java, Python, and JavaScript.

Как это работает

Amazon CodeWhisperer – это сервис на базе машинного обучения (ML), который помогает повысить производительность разработчиков, генерируя рекомендации по коду на основе их комментариев на естественном языке и кода в интегрированной среде разработки (IDE).

CodeWhisperer автоматически анализирует комментарии, определяет, какие облачные сервисы и публичные библиотеки лучше всего подходят для указанной задачи, и рекомендует фрагмент кода непосредственно в редакторе исходного кода. Рекомендации CodeWhisperer по коду основаны на моделях ML, обученных на различных источниках данных, включая Amazon и открытый исходный код. Разработчики могут принять верхнюю рекомендацию, просмотреть дополнительные рекомендации или продолжить писать свой собственный код.

CodeWhisperer обеспечивает проверку безопасности (для Java и Python), чтобы помочь разработчикам обнаруживать уязвимости в своих проектах и ответственно создавать приложения. Сервис также включает в себя систему отслеживания ссылок, которая определяет, может ли рекомендация по коду быть похожа на конкретные данные обучения. Затем разработчики могут легко найти и просмотреть пример кода и решить, использовать ли этот код в своем проекте. Кроме того, CodeWhisperer позволяет разработчикам избегать предвзятости, удаляя рекомендации по коду, которые могут быть сочтены предвзятыми и несправедливыми.

# Сравнение

Все рассмотренные продукты нацелены на то, чтобы ускорить разработку, каждый обладает функционалом дополнения кода, на основе контекста(Кроме Captain Stack). Из особенностей:

1. Copilot: превращение обычных человеческих слов и предложений в программный код.

2. Google ML-enhanced code completion: Распространяется сугубо среди сотрудников Google, самый закрытый продукт из представленных.

3. Captain Stack: Имеет отличительный функционал, а именно преобразует запрос в код, написан одним интузиастом.

4. IntelliCode: Предоставляет удобный функционал для командной разработки.

5. Code Whisperer: Подсказывает какие библиотеку/моудли лучше использовать, подходит для парного программирования, обеспечивает проверку безопасности.
